```python
import numpy as np 
import matplotlib.pyplot as plt
```

### Sigmoid

```python
sigmoid = lambda x: 1 / (1 + np.exp(-x))
```

```python
x=np.linspace(-10,10,10)
```

```python
y=np.linspace(-10,10,100)
```

```python
fig = plt.figure()
plt.plot(y,sigmoid(y),'b', label='linspace(-10,10,100)')
```

```python
plt.grid(linestyle='--')
```

```python
plt.xlabel('X Axis')
```

```python
plt.ylabel('Y Axis')
```

```python
plt.title('Sigmoid Function')
```

```python
plt.xticks([-4, -3, -2, -1, 0, 1, 2, 3, 4])
plt.yticks([-2, -1, 0, 1, 2])
```

```python
plt.ylim(-2, 2)
plt.xlim(-4, 4)
```

```python
plt.show()
#plt.savefig('sigmoid.png')
```

```python
fig = plt.figure()
```

### TanH

```python
tanh = lambda x: 2*sigmoid(2*x)-1
```

```python
x=np.linspace(-10,10,10)
```

```python
y=np.linspace(-10,10,100)
```

```python
plt.plot(y,tanh(y),'b', label='linspace(-10,10,100)')
```

```python
plt.grid(linestyle='--')
```

```python
plt.xlabel('X Axis')
```

```python
plt.ylabel('Y Axis')
```

```python
plt.title('TanH Function')
```

### ReLU

```python
relu = lambda x: np.where(x>=0, x, 0)
```

```python
x=np.linspace(-10,10,10)
```

```python
y=np.linspace(-10,10,1000)
```

```python
plt.plot(y,relu(y),'b', label='linspace(-10,10,100)')
```

```python
plt.grid(linestyle='--')
```

```python
plt.xlabel('X Axis')
```

```python
plt.ylabel('Y Axis')
```

```python
plt.title('ReLU')
```

```python
plt.xticks([-4, -3, -2, -1, 0, 1, 2, 3, 4])
plt.yticks([-4, -3, -2, -1, 0, 1, 2, 3, 4])
```

### Leaky ReLU

```python
leakyrelu = lambda x: np.where(x>=0, x, 0.1*x)
```

```python
x=np.linspace(-10,10,10)
```

```python
y=np.linspace(-10,10,1000)
```

```python
plt.plot(y,leakyrelu(y),'b', label='linspace(-10,10,100)')
```

```python
plt.grid(linestyle='--')
```

```python
plt.xlabel('X Axis')
```

```python
plt.ylabel('Y Axis')
```

```python
plt.title('Leaky ReLU')
```

### Binary Step

```python
bstep = lambda x: np.where(x>=0, 1, 0)
```

```python
x=np.linspace(-10,10,10)
```

```python
y=np.linspace(-10,10,1000)
```

```python
plt.plot(y,bstep(y),'b', label='linspace(-10,10,100)')
```

```python
plt.grid(linestyle='--')
```

```python
plt.xlabel('X Axis')
```

```python
plt.ylabel('Y Axis')
```

```python
plt.title('Step Function')
```

```python
plt.xticks([-4, -3, -2, -1, 0, 1, 2, 3, 4])
plt.yticks([-2, -1, 0, 1, 2])
```

```python
plt.ylim(-2, 2)
plt.xlim(-4, 4)
```

```python
plt.show()
#plt.savefig('step.png')
```

1) Design model (input, output, forward pass with different layers)<br>
2) Construct loss and optimizer<br>
3) Training loop<br>
      - Forward = compute prediction and loss<br>
      - Backward = compute gradients<br>
      - Update weights

```python
import torch
import torch.nn as nn
```

Linear regression<br>
f = w * x 

here : f = 2 * x

0) Training samples, watch the shape!

```python
X = torch.tensor([[1], [2], [3], [4]], dtype=torch.float32)
Y = torch.tensor([[2], [4], [6], [8]], dtype=torch.float32)
```

```python
n_samples, n_features = X.shape
print(f'#samples: {n_samples}, #features: {n_features}')
# 0) create a test sample
X_test = torch.tensor([5], dtype=torch.float32)
```

1) Design Model, the model has to implement the forward pass!<br>
Here we can use a built-in model from PyTorch

```python
input_size = n_features
output_size = n_features
```

we can call this model with samples X

```python
model = nn.Linear(input_size, output_size)
```


<br>
class LinearRegression(nn.Module):<br>
    def __init__(self, input_dim, output_dim):<br>
        super(LinearRegression, self).__init__()<br>
        # define diferent layers<br>
        self.lin = nn.Linear(input_dim, output_dim)<br>
    def forward(self, x):<br>
        return self.lin(x)<br>
model = LinearRegression(input_size, output_size)<br>


```python
print(f'Prediction before training: f(5) = {model(X_test).item():.3f}')
```

2) Define loss and optimizer

```python
learning_rate = 0.01
n_iters = 100
```

```python
loss = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

3) Training loop

```python
for epoch in range(n_iters):
    # predict = forward pass with our model
    y_predicted = model(X)

    # loss
    l = loss(Y, y_predicted)

    # calculate gradients = backward pass
    l.backward()

    # update weights
    optimizer.step()

    # zero the gradients after updating
    optimizer.zero_grad()
    if epoch % 10 == 0:
        [w, b] = model.parameters() # unpack parameters
        print('epoch ', epoch+1, ': w = ', w[0][0].item(), ' loss = ', l)
```

```python
print(f'Prediction after training: f(5) = {model(X_test).item():.3f}')
```
